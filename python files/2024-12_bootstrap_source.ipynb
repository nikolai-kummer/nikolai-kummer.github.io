```python
import numpy as np
import matplotlib.pyplot as plt
import os

from scipy.stats import skewnorm

OUTPUT_PATH = './../images/02'
if not os.path.exists(OUTPUT_PATH):
    os.makedirs(OUTPUT_PATH)

# Set random seed for reproducibility
np.random.seed(42)

# Define parameters for the skewed normal distribution
a = 5  # Skewness parameter; positive value indicates right skew
loc1 = 20  # Location parameter (mean) for the first sample
scale1 = 5  # Scale parameter (standard deviation) for the first sample

loc2 = 21  # Location parameter (mean) for the second sample
scale2 = 5  # Scale parameter (standard deviation) for the second sample

# Generate samples
sample1 = skewnorm.rvs(a, loc=loc1, scale=scale1, size=40)
sample2 = skewnorm.rvs(a, loc=loc2, scale=scale2, size=30)

# Clip values to ensure they fall within the desired range
sample1 = np.clip(sample1, 10, 50)
sample2 = np.clip(sample2, 10, 50)

# Create an array of indices for the x-axis
x1 = np.arange(1, len(sample1) + 1)
x2 = np.arange(1, len(sample2) + 1)
```

```python
import matplotlib.pyplot as plt

# Assuming x1, x2, sample1, and sample2 are already defined

# Set the style and context for the plot
plt.style.use('seaborn-darkgrid')  # Apply a clean and professional style

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(12, 7))

# Plot the scatter plot with enhanced aesthetics
scatter1 = ax.scatter(x1, sample1, color='royalblue', edgecolor='black', s=100, alpha=0.7, label='Prior to Change (n=40)')
scatter2 = ax.scatter(x2 + len(sample1), sample2, color='tomato', edgecolor='black', s=100, alpha=0.7, label='After Change (n=30)')

# Add labels and title with increased font sizes for better readability
ax.set_title('Impact of Process Change on Measurements', fontsize=20, pad=20)
ax.set_xlabel('Sample Index', fontsize=16, labelpad=15)
ax.set_ylabel('Measurement Value', fontsize=16, labelpad=15)

# Customize tick parameters for clarity
ax.tick_params(axis='both', which='major', labelsize=14)

# Add a legend with a shadow effect
legend = ax.legend(frameon=True, fontsize=14, loc='upper left', shadow=True)
legend.get_frame().set_edgecolor('black')

# Add a grid with customized appearance
ax.grid(True, linestyle='--', linewidth=0.6, alpha=0.7)

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure with a high resolution suitable for blogs
plt.savefig(os.path.join(OUTPUT_PATH, 'process_plot.png'), dpi=300, bbox_inches='tight')

# Display the plot
plt.show()

```
